```python
import pandas as pd

patient = pd.read_csv("C:\\Users\\oht31\\Desktop\\grk\\[GRK]Statistical Test Patient Data.csv")
prescription = pd.read_csv("C:\\Users\\oht31\\Desktop\\grk\\[GRK] Statistical Test Prescription Data - [GRK] Statistical Test Prescription Data.csv")
```

```python
general_indices = prescription[prescription['action_code'] == 'General'].index

for idx in general_indices:
    prescription.at[idx, 'action_code'] = prescription.at[idx, 'patient_location']
```

```python
prescription = prescription.applymap(lambda x: x.replace('(Self)', '') if isinstance(x, str) else x)

import re
def remove_numbers(text):
    return re.sub(r'\d+', '', text)


prescription['action_code'] = prescription['action_code'].apply(remove_numbers)
```

```python
prescription = prescription.merge(patient[['Anonymous_id', 'cancer_type']], on='Anonymous_id', how='left')


print(prescription)
```

```python
total_count = len(prescription['action_code'])

# action_code 열에 있는 데이터의 빈도 수 파악
action_code_counts = prescription['action_code'].value_counts()

# 각 데이터의 비율 계산
action_code_ratios = action_code_counts / total_count

print("Action Code의 전체 데이터 대비 비율:")
print(action_code_ratios)
```

```python
top_15_action_codes = action_code_counts.head(15).index

# 빈도 상위 15개 데이터를 제외한 나머지 데이터를 '기타'로 변경
prescription.loc[~prescription['action_code'].isin(top_15_action_codes), 'action_code'] = '기타'

print(prescription)
```

```python
fixed_nodes = {
    '유방암':'유방외과 외래',
    '대장직장암': '대장항문외과 외래',
    '위암': '위장관외과 외래',
    '폐암': '폐식도외과 외래' '호흡기내과 외래',
    '담췌암': '간담도췌외과 외래'
}

# prescription 데이터에 cancer_type을 기반으로 고정된 노드 정보 추가
prescription['fixed_node'] = prescription['cancer_type'].map(fixed_nodes)

# 결과 출력
print(prescription)
```

```python
## 가중치 계산을 위한 데이터 불러오기
import pandas as pd
df = pd.read_excel("C:\\Users\\oht31\\Desktop\\dfdf0327.xlsx")
```

```python
df.head()
```

```python
fixed_nodes = df['fixed_node'].unique()

network_data = pd.DataFrame(columns=['Source', 'Target', 'Weight'])

for node in fixed_nodes:
    # 각 고정된 노드에 해당하는 데이터 추출
    subset = df[df['fixed_node'] == node]
    # action_code의 빈도 계산
    action_code_freq = subset['action_code'].value_counts().reset_index()
    action_code_freq.columns = ['Target', 'Weight']
    action_code_freq['Source'] = node
    # 데이터를 network_data에 추가
    network_data = pd.concat([network_data, action_code_freq], ignore_index=True)


print(network_data)
```

```python
network_data.to_excel("C:\\Users\\oht31\\Desktop\\node2.xlsx")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 생성
G = nx.from_pandas_edgelist(network_data, 'Source', 'Target', edge_attr='Weight')

# 소스 노드 크기 설정
sizes = [3000 if node in network_data['Source'].unique() else 2000 for node in G.nodes]

# 소스와 타겟 색상 설정
node_colors = ['lightgreen' if node in network_data['Source'].unique() else 'lightblue' for node in G.nodes]

# 연결 선의 두께 설정
widths = [0.00005 * G[u][v]['Weight'] for u, v in G.edges()]

# 그래프 시각화
plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_size=sizes, node_color=node_colors, font_family='Batang', font_size=8, font_weight='bold', width=widths)
plt.title('네트워크 시각화', fontsize=16)
plt.show()
```
