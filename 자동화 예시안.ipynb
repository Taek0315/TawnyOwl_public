```python
#직무 관련 재능 색칠 코드




def update_table_colors(table, cell_values):
    for r in range(len(table.rows)):
        for c in range(len(table.columns)):
            cell_text = table.cell(r, c).text
            if cell_text in cell_values:
                cell = table.cell(r, c)
                cell.fill.solid()
                cell.fill.fore_color.rgb = RGBColor(174, 148, 237)  # 빨간색으로 변경

excel_path = "C:\\Users\\Dylan\\Desktop\\auto_ppt\\report.xlsx"
workbook = load_workbook(excel_path)
sheet1_name = "Sheet1"
sheet1 = workbook[sheet1_name]

for i in range(224, 226):
    st_values = [sheet1[f'{column}{i}'].value for column in 'NOPQR']

    prs = Presentation(f"C:\\Users\\Dylan\\Desktop\\auto_ppt\\0816\\output_({i-1}).pptx")

    slide_number = 45
    table_index = 8
    slide = prs.slides[slide_number]
    table = slide.shapes[table_index].table

    update_table_colors(table, st_values)

    slide_number2 = 46
    table_index2 = 8
    slide2 = prs.slides[slide_number2]
    table2 = slide2.shapes[table_index2].table

    update_table_colors(table2, st_values)

    output_ppt_path = f"C:\\Users\\Dylan\\Desktop\\auto_ppt\\0816\\output_({i-1}).pptx"
    prs.save(output_ppt_path)
```
