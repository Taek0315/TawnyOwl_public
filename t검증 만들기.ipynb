```python
import numpy as np
from scipy import stats as sc
import pandas as pd
data = {'A':[3,4,5,2,1,2,3,4,5,2,1,4,5],
        'B':[1,2,3,2,3,2,1,2,3,2,1,2,3],
        'C':[4,4,3,4,3,5,6,7,5,4,6,7,8]}
df = pd.DataFrame(data)

sample1 = df['A']
sample2 = df['B']
sample3 = df['C']

def t_test(sample1, sample2, sample3, alpha=0.05):
    mean1, mean2, mean3 = np.mean(sample1), np.mean(sample2), np.mean(sample3)
    std1,std2,std3 = np.std(sample1, ddof=1), np.std(sample2, ddof=1), np.std(sample3, ddof=3)
    n1,n2, n3 = len(sample1),len(sample2), len(sample3)

    means1 = mean1 - mean2
    means2 = mean1 - mean3
    means3 = mean2 - mean3
    de1 = np.sqrt((std1**2/n1)+(std2**2/n2))
    de2 = np.sqrt((std1**2/n1)+(std3**2/n3))
    de3 = np.sqrt((std2**2/n2)+(std3**2/n3))
    tvalue1 = means1/de1
    tvalue2 = means2/de2
    tvalue3 = means3/de3

    dof1 = n1 + n2 - 2
    dof2 = n1 + n3 - 2
    dof3 = n2 + n3 - 2
    pvalue1 = 2*(1-sc.t.cdf(abs(tvalue1), dof1))
    pvalue2 = 2*(1-sc.t.cdf(abs(tvalue2), dof2))
    pvalue3 = 2*(1-sc.t.cdf(abs(tvalue3), dof3))

    if pvalue1<alpha:
        result1 = '유의해용'
    else:
        result1 = '망했어용'
    
    if pvalue2<alpha:
        result2 = '유의해용'
    else:
        result2 = '망했어용'
    
    if pvalue3<alpha:
        result3 = '유의해용'
    else:
        result3 = '망했어용'
    
    if 0.01<pvalue1<=0.05:
        star1 = '한개에용'
    elif 0.001<pvalue1<=0.01:
        star1 = '두개에용'
    elif pvalue1<=0.001:
        star1 = '세개에용'
    else:
        star1 = '없어용'

    if 0.01<pvalue2<=0.05:
        star2='한개에용'
    elif 0.001<pvalue2<=0.01:
        star2='두개에용'
    elif pvalue3<=0.001:
        star2='세개에용'
    else:
        star3='없어용'
    
    if 0.01<pvalue3<=0.05:
        star3= '한개에용'
    elif 0.001<pvalue3<=0.01:
        star3='두개에용'
    elif pvalue3<=0.001:
        star3='세개에용'
    else:
        star3='없어용'

    result_df = pd.DataFrame({
        '검정 대상' : ['1집단-2집단', '1집단-3집단', '2집단-3진단'],
        't_value' : [round(tvalue1,2), round(tvalue2, 2), round(tvalue3, 3)],
        'p_value' : [round(pvalue1, 4), round(pvalue2, 4), round(pvalue3, 4)],
        '결과' : [result1, result2, result3],
        '별개수' : [star1, star2, star3]
    })
    return result_df

results_df = t_test(sample1, sample2, sample3)
print(results_df)
results_df.to_excel("C:\\Users\\oht31\\Desktop\\python\\t_test_result0311.xlsx")
# tvalue1, pvalue1, result1, star1, tvalue2,pvalue2,result2,star2,tvalue3,pvalue3,result3,star3 = t_test(sample1, sample2,sample3)
# print("1집단 2집단")
# print("t값_1:", round(tvalue1,2))
# print("p_value_1:",round(pvalue1,2))
# print("졸업가능유무_1:", result1)
# print("별개수_1:", star1)

# print("1집단, 3집단")
# print("t값_2:",round(tvalue2,2))
# print("p_value_2", round(pvalue2,2))
# print("졸업가능유무_2:", result2)
# print("별개수_2:",star2)

# print("2집단, 3집단")
# print("t값_3:", round(tvalue3 ,2))
# print("p_value_3:", round(pvalue3, 2))
# print("졸업가능유무_3:", result3)
# print("별개수_3:", star3)
```
